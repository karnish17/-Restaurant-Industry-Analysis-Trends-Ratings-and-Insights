```python
#Level 2
#Task 1: Table Booking and Online Delivery: Determine the percentage of restaurants that offer table booking and online delivery.
import pandas as pd
file_path="/content/drive/MyDrive/Copy of Dataset .csv"
df=pd.read_csv(file_path)
df
table_booking=df['Has Table booking'].value_counts(normalize=True).get('Yes')
online_delivery=df['Has Online delivery'].value_counts(normalize=True).get('Yes')
print(f"percentage of restaurant that offer table booking is {table_booking}")
print(f"percentage of restaurant that offer online delivery is {online_delivery}")
```

```python
#Level 3,Task 2
#Task: Customer Preference Analysis

#Analyze the relationship between the type of cuisine and the restaurant's rating.
import pandas as pd
import matplotlib.pyplot as plt

# Data Preparation
df['Cuisines'] = df['Cuisines'].fillna('Unknown').astype(str).str.split(', ')
df_exploded = df.explode('Cuisines')

# Analysis
kar6 = df_exploded.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head()

# Visualization
plt.figure(figsize=(8, 6))
plt.bar(kar6.index, kar6.values, color='red')
plt.title("Relationship between Cuisine and Restaurant's Rating")
plt.xlabel("Cuisines")
plt.ylabel("Aggregate Rating")
plt.xticks(rotation=45)
plt.show()



```

```python
#Level 3,Task 2
#Identify the most popular cuisines among customers based on the number of votes.
df.dropna(subset=['Cuisines'],inplace=True)

df['Cuisines']=df['Cuisines'].astype(str).str.split(',')
df_explode=df.explode('Cuisines')
kar6=df_explode.groupby('Cuisines')['Aggregate rating'].sum().sort_values(ascending=False).head()
kar6
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.title("most popular cuisines among customers based on the number of votes")
plt.xlabel("Cuisines")
plt.ylabel("Aggregate rating")
kar6.plot(kind='bar',colormap='viridis')
plt.grid(axis='y')
plt.show()
```
